```julia
using TikzGraphs
using TikzPictures # this is required for saving
using LightGraphs
```

```julia
g = DiGraph(4)
add_edge!(g, 1, 2)
add_edge!(g, 2, 3)
```

```julia
TikzGraphs.plot(g)
```

```julia
add_edge!(g, 3, 4)
add_edge!(g, 1, 4)
TikzGraphs.plot(g)
```

```julia
t = TikzGraphs.plot(g)
TikzPictures.save(PDF("./diagrams/graph"), t)
TikzPictures.save(SVG("./diagrams/graph"), t)
TikzPictures.save(TEX("graph"), t)
```

#### Edit g

```julia
TikzGraphs.plot(g, ["A", "B", "C", "D"])
```

```julia
TikzGraphs.plot(g, ["α", "β", "γ", "δ"])
```

```julia
TikzGraphs.plot(g, [L"\int_0^\infty f(x) dx",
                L"\sqrt{2}", 
                L"x^2",
                L"\frac{1}{2}"])
```

```julia
TikzGraphs.plot(g, ["α", "β", "γ", "α"])
```

```julia
TikzGraphs.plot(g, ["α", "β", "γ", "α"],
                node_style="draw, rounded corners, fill=blue!10")
```

```julia
TikzGraphs.plot(g, ["α", "β", "γ", "α"], 
    node_style="draw, rounded corners, fill=blue!10", 
    node_styles=Dict(1=>"fill=green!10",3=>"fill=yellow!10"))
```

### Editting edges

```julia
TikzGraphs.plot(g, ["α", "β", "γ", "α"],
    edge_labels=Dict((1,2)=>"2", (1,4)=>"4"))
```

```julia

TikzGraphs.plot(g, ["α", "β", "γ", "α"],
    edge_labels=Dict((1,2)=>"x", (1,4)=>"y"), 
    edge_style="green")
```

You can overide the style for specific edges as specified by a dictionary.



```julia
TikzGraphs.plot(g, 
    edge_labels=Dict((1,2)=>"x", (1,4)=>"y"),
    edge_style="green", 
    edge_styles=Dict((1,2)=>"blue"))
```

### Layouts

You can use different layouts (currently just Layered [default], Spring, and SimpleNecklace are supported).

```julia
TikzGraphs.plot(g, Layouts.Layered())
```

```julia
TikzGraphs.plot(g, Layouts.Spring())
```

```julia
TikzGraphs.plot(g, Layouts.Spring(randomSeed=52))
```

```julia
TikzGraphs.plot(g, Layouts.SimpleNecklace())
```

### More options

```julia
TikzGraphs.plot(g, 
    node_style="draw", 
    options="scale=2, font=\\huge\\sf")
```
